## Introduction

At Arcadia, we're interested in high-dimensional untargeted phenotyping that can be applied to a wide range of biological samples, from single cells to tissues. To this end, we’ve been using Raman spectroscopy to rapidly collect compositional and general chemical data on archaea, biomolecules, solvents, algae, and more. We’ve explored two modalities of this technique: spontaneous and coherent. Spontaneous Raman spectroscopy, which only requires one laser and a reasonably simple setup, is a relatively low-cost and valuable technique to probe the dynamic composition of biological samples. When placed on a microscope, you can do point-by-point mapping. You can acquire a full spectrum at each point, which is helpful if you’re taking a more agnostic approach or looking for multiple different peaks. This differs from spectral imaging with a coherent Raman system, which acquires images at specific vibrational frequencies using two synchronized laser beams. While coherent Raman microscopy typically has higher spatial resolution and can obtain a single image reasonably quickly, acquiring a full spectrum to search for peaks of interest takes longer. The two modalities can be used in a complementary manner. 

We have a DIY spontaneous Raman system, two commercial spontaneous Raman systems, and one coherent Raman confocal microscope in-house. To date, we’ve mainly been using spontaneous systems with simple focusing optics (lens or probe), but we have also explored adding them to a wide-field Nikon microscope. In this study, we had demos of two commercial confocal spontaneous Raman systems, which enable high-resolution mapping. We acquired data on samples of living *Chlamydomonas reinhardtii* and *Chlamydomonas smithii* cells. We tested three excitation wavelengths, 473 nm, 532 nm, and 785 nm; we already have Raman systems with the latter two in-house and are interested in adding another line in the UV or blue range. Here, we share an analysis notebook that we used to visualize and analyze spectral and map data from both commercial systems. Our intent with this study is to explore the types of scientific questions that could be enabled with this combination of spatial and spectral information on biological samples. 

We hope the notebook will be helpful to biologists interested in different types of Raman microscopy for their samples and to anyone interested in visualizing map data from these commercial systems.  

## Background

In this study, we looked at the composition of single *Chlamydomonas* cells with three different excitation wavelengths. *Chlamydomonas reinhardtii* and *Chlamydomonas smithii* are two species of interest for Arcadia Science and have been part of our organism selection efforts. *Chlamydomonas* is a genus of unicellular green algae that are photosynthetic, have flagella, and are about 10 μm in size [cite]. They contain cell components that have been studied using Raman spectroscopy [cite][cite]. For instance, the chloroplasts give clear pigment peaks (chlorophyll and carotenoids), the eyespot is apparent due to strong carotenoid (B-carotene) peaks [cite], and starch granules may be detectable using the 478 cm^-1^ peak caused by C-C-C bending and C-O stretching modes [cite]. Both species have strong structural and compositional similarities; *C. reinhardtii* is the more common model organism. 

The excitation wavelength used in spontaneous Raman is an important factor in the data you can acquire because light of different energies interacts with samples differently. The scattering efficiency of the light decreases as the wavelength increases, meaning that the same compound might have weaker peaks when excited with 785 nm versus 532 nm. However, excitation wavelengths in the UV or IR can avoid background fluorescence, especially with biological samples. Finally, resonance enhancement effects can come into play when the excitation wavelength is close to an electronic transition of compounds in the sample. This yields a more intense signal for those specific compounds. For instance, carotenoids have an absorption band — electronic transition — between 450-520 nm, so they're resonantly enhanced under 473 nm and 532 nm (Figure 1).  

Our goals were to understand:

* If we could use spontaneous Raman mapping to detect distinct biological structures (e.g., eyespot) 
* Which excitation wavelength yielded the most resolvable features, and if they had complementary information
* What data analysis methods are useful for spatially resolved Raman data.

# UV-Vis absorbance of pigment standards

The pigment absorbance spectra show where each molecule absorbs light most strongly, which predicts where resonance Raman enhancement may occur. Carotenoids (β-carotene, lycopene) absorb broadly between ~420–520 nm, with peak absorption near 450–480 nm. Both the 473 nm and 532 nm laser lines fall within or near this window, predicting resonance enhancement of carotenoid Raman signals under both excitation wavelengths. The 473 nm line sits closer to the β-carotene absorption maximum, suggesting potentially stronger enhancement. Chlorophylls a and b have Soret absorption bands near 430–435 nm and Q-band absorption near 640–680 nm. None of the three excitation wavelengths closely overlap these transitions, so chlorophyll peaks are not expected to be resonantly enhanced. The 785 nm laser is well outside the absorption bands of all four pigments, so all Raman signals at this wavelength reflect non-resonant scattering and are expected to be weaker relative to carotenoid signals seen at the shorter wavelengths.


```python
# | code-fold: true
# | code-summary: "Source code for this figure:"
# | label: fig-uvvis-all
# | fig-cap: "Figure 2: UV-Vis spectra of pigment standards in acetone."

import arcadia_pycolor as apc
import pandas as pd
import plotly.graph_objects as go

apc.plotly.setup()

# Step 1: Read the TSV manually
def import_tsv(file_path):
    with open(file_path) as f:
        lines = [line.strip().split('\t') for line in f if line.strip()]
    df = pd.DataFrame(lines)
    return df

# Step 2: Extract sample blocks and names
def split_uv_vis_data(df):
    split_indices = df.index[df.iloc[:, 0] == 'Wavelength (nm)'].tolist()
    split_dfs = []
    labels = []

    for i in range(len(split_indices)):
        header_idx = split_indices[i]
        sample_name = df.iloc[header_idx - 2, 0]  # sample name is two rows above header
        start_idx = header_idx + 1
        end_idx = split_indices[i + 1] - 3 if i + 1 < len(split_indices) else len(df)

        headers = df.iloc[header_idx]
        data = df.iloc[start_idx:end_idx + 1].copy()
        data.columns = headers
        split_dfs.append(data)
        labels.append(sample_name)

    return split_dfs, labels

# Step 3a: Combined plot with Plotly
figure_size = apc.mpl.get_figure_dimensions("float")
def plot_all_spectra_plotly(dfs, labels):
    fig = go.Figure()
    for df, label in zip(dfs, labels, strict=False):
        df = df.dropna()
        df = df.apply(pd.to_numeric, errors='coerce')
        fig.add_trace(go.Scatter(
            x=df['Wavelength (nm)'],
            y=df.iloc[:, 1],
            mode='lines',
            name=label,
        ))

    fig.update_layout(
        xaxis_title="Wavelength (nm)",
        yaxis_title="Absorbance",
        xaxis_range=[350, 800],
        paper_bgcolor=apc.parchment,
        plot_bgcolor=apc.parchment,
        width=figure_size[0] * 100,
        height=figure_size[1] * 100,
    )
    apc.plotly.style_plot(fig, monospaced_axes="xy")
    fig.show()

# ---- Main Execution ----
file_path = 'data/UV-Vis 5_1_2025 7_19_21 PM.tsv'
uv_vis_raw = import_tsv(file_path)
split_uv_vis, sample_labels = split_uv_vis_data(uv_vis_raw)

plot_all_spectra_plotly(split_uv_vis, sample_labels)
```

## Initial spectral plot comparison

First, let's plot the data from each experiment to have a spectra acquired with 473 nm, 532 nm, and 785 nm excitation side by side. This will help us visually identify similarities and differences between the resonance effects and find specific Raman shift values that could be biological markers.

We'll start with 473 nm data from the Horiba LabRam Evolution. This map has 1073 points and covers four cells and white space between and around them. We can apply a gentle Savitzky-Golay or median filter to remove some of the noise, which can be due to detector noise or cosmic rays. Both options are in the cell below, and you can comment in the filter of your choice. Each figure in this section has median filtered Raman data (kernel size = 5). 

```python
# | code-fold: true
# | code-summary: "Source code for this figure:"
# | label: fig-473-all
# | fig-cap: "1073 Raman spectra of *Chlamydomonas reinhardtii* CC-124 cells with 473 nm excitation. Each trace is a single point on the map."

import matplotlib.pyplot as plt
import scipy.signal
from pybaselines.whittaker import airpls

apc.mpl.setup()

# Load Horiba 473 data

blue_path = "data/Cells- 473nm-Map-ForPeng 1.txt"
blue_data = pd.read_csv(blue_path, sep="\t")

# Rename first two columns to x and y positions

blue_data = blue_data.rename(columns={"Unnamed: 0": "x_pos", "Unnamed: 1": "y_pos"})

# Separate the data into two dataframes, one for xy positions and one for intensity values

blue_xy = blue_data[["x_pos", "y_pos"]]
blue_intensity = blue_data.drop(columns=["x_pos", "y_pos"])
blue_num_points = len(blue_xy)

# Flip the intensity dataframe horizontally and convert column names to float

blue_intensity = blue_intensity.iloc[:, ::-1]
blue_intensity.columns = blue_intensity.columns.astype(float)

# Crop the intensity data to only include Raman shifts between 300 and 3600 cm-1

blue_intensity = blue_intensity.loc[:, (blue_intensity.columns >= 300)
                                    & (blue_intensity.columns <= 3600)]

# # Apply a Savitzky-Golay filter to the intensity data

# blue_intensity_filt = blue_intensity.apply(lambda x: scipy.signal.savgol_filter(
#     x, window_length=11, polyorder=5))

# Apply a median filter to the intensity data

blue_intensity_filt = blue_intensity.apply(lambda x: scipy.signal.medfilt(x, kernel_size=5))

# Plot spectra in blue_intensity_filt using matplotlib

figure_size = apc.mpl.get_figure_dimensions("float")
fig, ax = plt.subplots(
    figsize=figure_size,
    constrained_layout=True,
    #dpi=300,
    facecolor=apc.parchment
)
colors = apc.palettes.blue_shades.colors
for i in range(len(blue_intensity_filt)):
    ax.plot(blue_intensity_filt.columns, blue_intensity_filt.iloc[i],
            label=f"Point {i+1}", color=colors[i % len(colors)])
ax.set_xlabel("Raman shift (cm$^{-1}$)")
ax.set_ylabel("Intensity")
apc.mpl.style_plot(monospaced_axes="both")
ax.set_facecolor(apc.parchment)

plt.show()
```

The spectra show that the most intense peaks are at ~1150, ~1520, and ~2650 cm^-1^. Some spectra have minor peaks at ~1005, between 2200-2500, and ~3100 cm^-1^. Even with no baseline removal applied to the spectra, there's only a bit of fluorescence, which causes the baseline to rise. It's also apparent that some points have no signal, which is reasonable considering many positions didn't hit a cell. Let's make a similar plot with the 532 nm data from the Horiba LabRam Evolution, which we acquired on another group of cells from the same slide, with a differently sized map (873 points).

```python
# | code-fold: true
# | code-summary: "Source code for this figure:"
# | label: fig-532-all
# | fig-cap: "837 Raman spectra of *Chlamydomonas reinhardtii* CC-124 cells with 532 nm excitation. Each trace is a single point on the map."

# Load Horiba 532 data

green_path = "data/Cells- 532nm-Map-ForPeng 1.txt"
green_data = pd.read_csv(green_path, sep="\t")

# Rename first two columns to x and y positions

green_data = green_data.rename(columns={"Unnamed: 0": "x_pos", "Unnamed: 1": "y_pos"})

# Separate the data into two dataframes, one for xy positions and one for intensity values

green_xy = green_data[["x_pos", "y_pos"]]
green_intensity = green_data.drop(columns=["x_pos", "y_pos"])
green_num_points = len(green_xy)

# Flip the intensity dataframe horizontally and convert column names to float

green_intensity = green_intensity.iloc[:, ::-1]
green_intensity.columns = green_intensity.columns.astype(float)

# Crop the intensity data to only include Raman shifts between 300 and 3600 cm-1

green_intensity = green_intensity.loc[:, (green_intensity.columns >= 300) & (green_intensity.columns <= 3600)]

# # Apply a Savitzky-Golay filter to the intensity data

# green_intensity_filt = green_intensity.apply(lambda x: scipy.signal.savgol_filter(x, window_length=35, polyorder=3))

# Apply a median filter to the intensity data

green_intensity_filt = green_intensity.apply(lambda x: scipy.signal.medfilt(x, kernel_size=5))

# Plot spectra in green_intensity using matplotlib; x values are the header row of green_intensity

figure_size = apc.mpl.get_figure_dimensions("float")
fig, ax = plt.subplots(
    figsize=figure_size,
    constrained_layout=True,
    # dpi=300,
    facecolor=apc.parchment
)
colors = apc.palettes.green_shades.colors
for i in range(len(green_intensity_filt)):
    ax.plot(green_intensity_filt.columns, green_intensity_filt.iloc[i], label=f"Point {i+1}", color=colors[i % len(colors)])
ax.set_xlabel("Raman shift (cm$^{-1}$)")
ax.set_ylabel("Intensity")
apc.mpl.style_plot(monospaced_axes="both")
ax.set_facecolor(apc.parchment)

plt.show()
```

Looking at the spectra, it seems that the most intense peaks are at ~1005, ~1150, and ~1520 cm^-1^. It's hard to discern minor peaks, though there's possibly a shoulder on the 1150 peak. In comparison to the 473 nm data, there's more fluorescence, with a tail increasing >2500 cm^-1^. 

Let's make a similar plot with the 785 nm data from the Renishaw inVia Qontor. In this case, we mapped a single *Chlamydomonas smithii* CC-1373 cell with a more limited spectral range for a quicker acquisition. 

```python
# | code-fold: true
# | code-summary: "Source code for this figure:"
# | label: fig-785-all
# | fig-cap: "285 Raman spectra of *Chlamydomonas smithii* CC-1373 cells with 785 nm excitation. Each trace is a single point on the map."

import numpy as np

# Load Renishaw 785 data - LONG FORMAT
red_path = "data/2024-11-12_CC-1373_TAP_slide_10_singlecellmap.txt"

# Read with only the first 4 columns (X, Y, Wave, Intensity)
df = pd.read_csv(red_path, sep="\t", usecols=[0, 1, 2, 3],
                 names=['X', 'Y', 'Wave', 'Intensity'], skiprows=1)

# Get unique Raman shifts
raman_shifts = sorted(df['Wave'].unique())

# Pivot to wide format: rows = spatial points, columns = Raman shifts
data_wide = df.pivot_table(index=['X', 'Y'], columns='Wave', values='Intensity', aggfunc='first')
data_wide = data_wide.reset_index()

# Extract spatial positions AFTER pivoting to ensure correct order
red_xy = data_wide[['X', 'Y']].copy()
red_num_points = len(red_xy)

# Extract intensity columns
red_intensity_restructured = data_wide.drop(columns=['X', 'Y'])
red_intensity_restructured.columns = red_intensity_restructured.columns.astype(float)

# Flip to match Horiba orientation (high to low wavenumber)
red_intensity_restructured = red_intensity_restructured.iloc[:, ::-1]

# Apply a median filter to the intensity data
red_intensity_filt = red_intensity_restructured.apply(lambda x: scipy.signal.medfilt(x, kernel_size=5))

# Plot spectra using matplotlib
figure_size = apc.mpl.get_figure_dimensions("float")
fig, ax = plt.subplots(
    figsize=figure_size,
    constrained_layout=True,
    facecolor=apc.parchment
)
colors = apc.palettes.red_shades.colors
for i in range(len(red_intensity_filt)):
    ax.plot(red_intensity_filt.columns, red_intensity_filt.iloc[i],
            label=f"Point {i+1}", color=colors[i % len(colors)])
ax.set_xlabel("Raman shift (cm$^{-1}$)")
ax.set_ylabel("Intensity")
apc.mpl.style_plot(monospaced_axes="both")
ax.set_facecolor(apc.parchment)

plt.show()
```

Looking at the 785 spectra, there's a clear background fluorescence bump around 1400 cm^-1^ due to the light interacting with the glass slide and some background fluorescence from the cells. You can still see peaks at ~1150 and ~1520 cm^-1^, though they might be less intense than in the other wavelengths.  

Even without applying baseline correction or normalization, it's clear that we can see some real Raman signals in each dataset. Across all three wavelengths, we see some common peaks at ~1157 and 1520 cm^-1^. This combination of peaks matches well with C-C stretching, and C=C stretching modes of carotenoids [cite], which would be expected in photosynthetic algae [cite]. The 473 nm spectra certainly have the greatest number of peaks distinguishable by eye, whereas the 532 nm spectra have fewer, and the 785 nm fewer still. This is another reason to consider that carotenoids are responsible for some of the peaks - they absorb 473 nm light well (Figure 1), which enables resonance enhancement and increased signal [cite].

## Initial spectral map comparison

To better explore the data and find spatial hot spots, we can present the data differently. Each
file contains x, y positional data for the point at which a spectrum was taken, which can be put
on a plot. Then the user can scroll through the Raman shift space to see the intensity of each map
point at a specific wavenumber. 

The size of the points is not the same as the beam size of the spectrometer, nor is the spacing 
accurate. 

```python
# | code-fold: true
# | code-summary: "Source code for this figure:"
# | label: fig-473spectralmap
# | fig-cap: "Interactive spectral map of *Chlamydomonas reinhardtii* CC-124 cells with 473 nm excitation."

import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Read in data
with open(blue_path) as f:
    header_line = f.readline().strip().split()
raman_shifts = [float(val) for val in header_line]

# Read the wide-format data:
# First two columns: x and y positions;
# subsequent columns: intensity values corresponding to each Raman shift
data = pd.read_csv(blue_path, sep=r'\s+', skiprows=1, header=None)
data = data.rename(columns={0: 'x', 1: 'y'})

# Rename intensity columns: column i+2 corresponds to raman_shifts[i]
intensity_columns = {i+2: raman_shifts[i] for i in range(len(raman_shifts))}
data = data.rename(columns=intensity_columns)

# Build allowed_cols as a list of Raman shift values that fall within 300–3600 cm⁻¹.
allowed_cols = [s for s in raman_shifts if 300 <= s <= 3600]

# Melt only the allowed intensity columns to build a long-format DataFrame
df_long = data.melt(id_vars=['x', 'y'], value_vars=allowed_cols,
                    var_name='raman_shift', value_name='intensity')
df_long['raman_shift'] = df_long['raman_shift'].astype(float)
df_long['intensity'] = pd.to_numeric(df_long['intensity'], errors='coerce')
df_long = df_long.sort_values(by='raman_shift')

# Compute the mean spectrum from the cropped (allowed) data
mean_spec = df_long.groupby('raman_shift', as_index=False)['intensity'].mean()

# Update unique Raman shifts from the cropped data.
unique_shifts = sorted(df_long['raman_shift'].unique())

# Convert colorscale ONCE before creating frames (performance optimization)
magma_colorscale = apc.gradients.magma.to_plotly_colorscale()

# Create a combined figure with two rows:
# Row 1: Spectral map
# Row 2: Mean Raman spectrum
fig = make_subplots(rows=2, cols=1,
                    row_heights=[0.65, 0.35],
                    vertical_spacing=0.15)
                    # subplot_titles=["Spectral map of Chlamydomonas cells (473 nm)", "Mean spectrum"])

# For the spectral map, use the original x,y (preserving order).
x_full = data['x']
y_full = data['y']

# --- Initial Frame ---
init_shift = unique_shifts[0]
# The initial allowed column is simply the value of init_shift.
init_col = init_shift
intensity_init = data[init_col]

# Spectral map trace (top panel)
map_trace = go.Scatter(
    x=x_full,
    y=y_full,
    mode='markers',
    marker=dict(
        size=10,
        color=intensity_init,
        colorscale=magma_colorscale,
        colorbar=dict(title=dict(text='Intensity'),
                      tickfont=dict(color='black'))
    ),
    name="Spectral map of Chlamydomonas cells (473 nm)"
)
fig.add_trace(map_trace, row=1, col=1)

# Add mean spectrum trace (bottom panel)
mean_trace = go.Scatter(
    x=mean_spec['raman_shift'],
    y=mean_spec['intensity'],
    mode='lines',
    line=dict(color='black'),
    name="Mean spectrum"
)
fig.add_trace(mean_trace, row=2, col=1)

# Add an initial vertical indicator on the mean spectrum if init_shift != 0
if init_shift != 0:
    init_line = dict(
        type='line',
        x0=init_shift,
        x1=init_shift,
        y0=mean_spec['intensity'].min(),
        y1=mean_spec['intensity'].max(),
        line=dict(color="black", width=2),
        opacity=0.5,
        xref="x2",  # reference to the x-axis of subplot row 2
        yref="y2"
    )
    fig.update_layout(shapes=[init_line])

# Create animation frames
frames = []
for shift in unique_shifts:
    intensity_vals = data[shift]
    shapes_list = []
    if shift != 0:
        shapes_list = [dict(
            type='line',
            x0=shift,
            x1=shift,
            y0=mean_spec['intensity'].min(),
            y1=mean_spec['intensity'].max(),
            line=dict(color="black", width=2),
            opacity=0.5,
            xref="x2",
            yref="y2"
        )]
    frame = go.Frame(
        data=[
            go.Scatter(
                x=x_full,
                y=y_full,
                marker=dict(color=intensity_vals, colorscale=magma_colorscale)
            ),
            go.Scatter(
                x=mean_spec['raman_shift'],
                y=mean_spec['intensity'],
                mode='lines',
                line=dict(color='black')
            )
        ],
        layout=dict(shapes=shapes_list),
        name=str(shift)
    )
    frames.append(frame)
fig.frames = frames

# Slider Configuration
slider_steps = []
for shift in unique_shifts:
    step = dict(
        method="animate",
        args=[
            [str(shift)],
            dict(mode="immediate", frame=dict(duration=100, redraw=True), transition=dict(duration=0))
        ],
        label=str(round(shift, 1))
    )
    slider_steps.append(step)

fig.update_layout(
    sliders=[dict(
        active=0,
        currentvalue=dict(prefix="Raman shift: "),
        pad={"t": 10},
        steps=slider_steps,
        x=0,
        y=-0.1,
        len=1,
        # font=dict(color='black', family="Helvetica Neue")
    )],
    width=1000,
    height=1200,
    showlegend=False,
    plot_bgcolor='black',  # This applies to all subplots by default.
    paper_bgcolor='#FDF8F2',  # Overall paper color.
    font=dict(family="Helvetica Neue", color="black")
)

# Axis Styling for Spatial Map (Row 1)
x_min = data['x'].min()
x_max = data['x'].max()
y_min = data['y'].min()
y_max = data['y'].max()

fig.update_xaxes(row=1, col=1, showgrid=False, showline=True, linewidth=1, linecolor='black',
                 title="X (μm)",
                 range=[x_min, x_max])
fig.update_yaxes(row=1, col=1, showgrid=False, showline=True, linewidth=1, linecolor='black',
                 title="Y (μm)",
                 range=[y_min, y_max])

# Axis Styling for Mean Spectrum (Row 2)
fig.update_xaxes(row=2, col=1, showgrid=False, showline=True, linewidth=1, linecolor='black',
                 title="Raman shift (cm⁻¹)")
fig.update_yaxes(row=2, col=1, showgrid=False, showline=True, linewidth=1, linecolor='black',
                 title="Mean intensity")

# Enforce equal aspect ratio for the spatial map (Row 1)
fig.update_xaxes(row=1, col=1, scaleanchor="y", scaleratio=1)

# --- Set a Custom Background for the Mean Spectrum Subplot ---
# Since subplot-specific background colors aren't directly settable,
# add a rectangle shape covering the domain of row 2.
# Get the domain for row 2 from xaxis2 and yaxis2
x2_domain = fig.layout.xaxis2.domain  # list: [start, end]
y2_domain = fig.layout.yaxis2.domain  # list: [start, end]
fig.add_shape(
    type="rect",
    xref="paper",
    yref="paper",
    x0=x2_domain[0],
    x1=x2_domain[1],
    y0=y2_domain[0],
    y1=y2_domain[1] + 0.02,
    fillcolor="#FDF8F2",
    layer="below",
    line_width=0,
)

apc.plotly.style_plot(fig, monospaced_axes="all")
apc.plotly.style_legend(fig)
# apc.plotly.style_plot(fig, colorbar_exists=True)
apc.plotly.add_commas_to_axis_tick_labels(fig)
fig.show()

# # Save the figure as HTML
# fig.write_html("fig-473-heatmap-thermal.html")
```

## Data refinement

While the initial spectral plots and maps are already illuminating, two refinement steps improve the quality of downstream analysis. First, intensity-based spatial masking removes background points — map positions that did not intersect a cell — so that subsequent analysis is not diluted by off-cell signal. Second, baseline correction removes broad fluorescence contributions that ride underneath the Raman peaks, making peak positions and relative intensities more accurate and comparable across wavelengths.


### Cell masking with intensity thresholding

To improve our analysis, we need to separate spatial points that are "on-cell" from background points. We'll use simple intensity thresholding based on mean intensity across the spectrum. This approach is interpretable and works well when cells have consistently higher signal than background. Thresholds were set at the 25th, 30th, and 40th percentile of mean intensity for the 473 nm, 532 nm, and 785 nm datasets respectively, chosen by visual inspection of the resulting masks to retain clearly on-cell points while excluding background. The higher threshold for the 785 nm dataset reflects its stronger fluorescence background, which raises the mean intensity of background points relative to the other datasets.

### Baseline correction and peak analysis

Now that we have masked data containing only "on-cell" points, let's apply baseline correction to remove fluorescence backgrounds and identify peaks. We'll use the airPLS algorithm from pybaselines, which is effective at removing broad baseline features while preserving sharp spectral peaks.


```python
# Baseline correction for all three wavelengths

# Define wavenumber arrays from mean spectrum indices (needed for downstream analysis)
wavenumbers_473 = mean_473.index.values
wavenumbers_532 = mean_532.index.values
wavenumbers_785 = mean_785.index.values

# 473nm and 532nm: airPLS
baseline_result_473 = airpls(mean_473.values, lam=1e6)
baseline_473 = baseline_result_473[0] if isinstance(baseline_result_473, tuple) else baseline_result_473
corrected_473 = mean_473.values - baseline_473

baseline_result_532 = airpls(mean_532.values, lam=1e6)
baseline_532 = baseline_result_532[0] if isinstance(baseline_result_532, tuple) else baseline_result_532
corrected_532 = mean_532.values - baseline_532
```

```python
# | code-fold: true
# | code-summary: "Source code for improved 785nm baseline correction:"
# | label: fig-785-background-subtraction
# | fig-cap: "Background subtraction prior to baseline correction for 785 nm data."

# ========== 785nm BASELINE CORRECTION WITH BACKGROUND SUBTRACTION ==========

# Step 1: Identify background (off-cell) points using the mask
background_points_785 = red_intensity_filt[~mask_785]

# Calculate mean background spectrum from off-cell points
mean_background_785 = background_points_785.mean(axis=0)

# Step 2: Subtract background from all cell spectra BEFORE baseline correction
# This removes the bulk glass fluorescence
cell_785_bkgd_subtracted = cell_785 - mean_background_785.values

# Recalculate mean spectrum after background subtraction
mean_785_bkgd_subtracted = cell_785_bkgd_subtracted.mean(axis=0)

# Step 3: Apply airPLS baseline correction to the background-subtracted data
baseline_result_785 = airpls(mean_785_bkgd_subtracted, lam=1e6)
baseline_785 = baseline_result_785[0] if isinstance(baseline_result_785, tuple) else baseline_result_785
corrected_785 = mean_785_bkgd_subtracted - baseline_785

# Step 4: Create figure showing the process
figure_size = apc.mpl.get_figure_dimensions("full_wide")
fig, axes = plt.subplots(
    figsize=figure_size,
    layout="constrained",
    nrows=1,
    ncols=1,
    facecolor=apc.parchment
)

# Panel 1: Background spectrum and subtraction
axes.plot(wavenumbers_785, mean_785.values, color=apc.bark, alpha=0.3, linewidth=2, label='Mean cell')
axes.plot(mean_background_785.index.values, mean_background_785.values,
             color=apc.bark, linestyle='--', linewidth=2, label='Mean background')
axes.plot(wavenumbers_785, mean_785_bkgd_subtracted, color=apc.mars, linewidth=2, label='Background-subtracted')
axes.set_xlabel('Raman shift (cm$^{-1}$)')
axes.set_ylabel('Intensity')
axes.legend(loc='upper right')
axes.grid(False)

apc.mpl.style_plot(monospaced_axes="both")
plt.savefig('assets/fig-785-background-subtraction.pdf', dpi=300, bbox_inches='tight')
plt.show()
```

```python
# | code-fold: true
# | code-summary: "Source code for manual peak detection and SNR analysis:"
# | label: fig-peak-detection-snr
# | fig-cap: "Manual peak detection with SNR calculation for all three wavelengths"

import numpy as np
from scipy.signal import find_peaks

# ========== MANUAL PEAK DETECTION ==========

# Ground truth peaks manually identified by user
manual_peaks_473 = np.array([476, 518, 608, 704, 721, 746, 762, 795, 763, 841, 922, 962, 1008, 1130, 1162, 1193, 1214, 1232, 1269, 1295, 1331, 1352, 1359, 1395, 1421, 1444, 1530, 1558, 1568, 1606, 1636, 1641, 2019, 2162, 2317, 2347, 2488, 2533, 2682, 2711, 3001, 3046, 3330, 3489])
manual_peaks_532 = np.array([520, 871, 965, 1007, 1157, 1188, 1212, 1269, 1285, 1353, 1393, 1452, 1524, 1605, 2163, 2310, 2528, 2676])
manual_peaks_785 = np.array([746, 755, 910, 985, 1005, 1053, 1111, 1142, 1157, 1187, 1228, 1286, 1328, 1442, 1523, 1554])

# Normalize spectra
norm_473 = corrected_473 / corrected_473.max()
norm_532 = corrected_532 / corrected_532.max()
norm_785 = corrected_785 / corrected_785.max()

# Find closest peak in spectrum for each manual peak
def find_closest_peak(spectrum, wavenumbers, target_wavenumber, search_window=10):
    """
    Find the actual peak position closest to the target wavenumber.
    Returns: (peak_index, peak_wavenumber, peak_height)
    """
    # Find indices within search window
    mask = np.abs(wavenumbers - target_wavenumber) < search_window
    if not np.any(mask):
        # If no points in window, just use closest point
        idx = np.argmin(np.abs(wavenumbers - target_wavenumber))
        return idx, wavenumbers[idx], spectrum[idx]

    # Find local maximum within window
    window_indices = np.where(mask)[0]
    window_spectrum = spectrum[mask]
    local_max_idx = np.argmax(window_spectrum)
    peak_idx = window_indices[local_max_idx]

    return peak_idx, wavenumbers[peak_idx], spectrum[peak_idx]

# Find actual peak positions and heights for manual peaks
peaks_473 = []
peak_wavenumbers_473 = []
peak_heights_473 = []
for target in manual_peaks_473:
    idx, wn, height = find_closest_peak(norm_473, wavenumbers_473, target)
    peaks_473.append(idx)
    peak_wavenumbers_473.append(wn)
    peak_heights_473.append(height)

peaks_473 = np.array(peaks_473)
peak_wavenumbers_473 = np.array(peak_wavenumbers_473)
peak_heights_473 = np.array(peak_heights_473)

peaks_532 = []
peak_wavenumbers_532 = []
peak_heights_532 = []
for target in manual_peaks_532:
    idx, wn, height = find_closest_peak(norm_532, wavenumbers_532, target)
    peaks_532.append(idx)
    peak_wavenumbers_532.append(wn)
    peak_heights_532.append(height)

peaks_532 = np.array(peaks_532)
peak_wavenumbers_532 = np.array(peak_wavenumbers_532)
peak_heights_532 = np.array(peak_heights_532)

peaks_785 = []
peak_wavenumbers_785 = []
peak_heights_785 = []
for target in manual_peaks_785:
    idx, wn, height = find_closest_peak(norm_785, wavenumbers_785, target)
    peaks_785.append(idx)
    peak_wavenumbers_785.append(wn)
    peak_heights_785.append(height)

peaks_785 = np.array(peaks_785)
peak_wavenumbers_785 = np.array(peak_wavenumbers_785)
peak_heights_785 = np.array(peak_heights_785)

print(f"\n{'='*60}")
print("MANUAL PEAK DETECTION RESULTS")
print(f"{'='*60}\n")
print(f"473 nm: {len(peaks_473)} peaks")
print(f"532 nm: {len(peaks_532)} peaks")
print(f"785 nm: {len(peaks_785)} peaks")
print(f"\n473 nm peaks (cm⁻¹): {', '.join([f'{w:.0f}' for w in peak_wavenumbers_473])}")
print(f"\n532 nm peaks (cm⁻¹): {', '.join([f'{w:.0f}' for w in peak_wavenumbers_532])}")
print(f"\n785 nm peaks (cm⁻¹): {', '.join([f'{w:.0f}' for w in peak_wavenumbers_785])}")

# ========== SNR CALCULATION ==========

# Estimate noise from quiet regions (wavelength-specific)
# 473nm and 532nm: 2000-2100 cm⁻¹ (free of biological signals)
# 785nm: 1700-1800 cm⁻¹ (different quiet region)
noise_region_473 = (wavenumbers_473 > 2000) & (wavenumbers_473 < 2100)
noise_std_473 = np.std(norm_473[noise_region_473])

noise_region_532 = (wavenumbers_532 > 2000) & (wavenumbers_532 < 2100)
noise_std_532 = np.std(norm_532[noise_region_532])

noise_region_785 = (wavenumbers_785 > 1700) & (wavenumbers_785 < 1800)
noise_std_785 = np.std(norm_785[noise_region_785])

# Calculate SNR for each detected peak
snr_473 = peak_heights_473 / noise_std_473
snr_532 = peak_heights_532 / noise_std_532
snr_785 = peak_heights_785 / noise_std_785

print(f"\n{'='*60}")
print("SIGNAL-TO-NOISE RATIOS")
print(f"{'='*60}\n")
print("Noise region for 473nm & 532nm: 2000-2100 cm⁻¹")
print("Noise region for 785nm: 1700-1800 cm⁻¹")
print(f"\n473 nm: Mean SNR = {snr_473.mean():.1f} ± {snr_473.std():.1f} (range: {snr_473.min():.1f} - {snr_473.max():.1f})")
print(f"532 nm: Mean SNR = {snr_532.mean():.1f} ± {snr_532.std():.1f} (range: {snr_532.min():.1f} - {snr_532.max():.1f})")
print(f"785 nm: Mean SNR = {snr_785.mean():.1f} ± {snr_785.std():.1f} (range: {snr_785.min():.1f} - {snr_785.max():.1f})")

# ========== VISUALIZATION ==========
# Peak markers are colored by SNR using a sequential colormap.
# High-SNR peaks (warmer color) are reliably above noise; low-SNR peaks (cooler)
# may be marginal. The colorbar on each panel shows the SNR scale.

import matplotlib.cm as cm
import matplotlib.colors as mcolors

fig, axes = plt.subplots(3, 1, figsize=(14, 12), layout="constrained", facecolor=apc.parchment)

datasets = [
    (axes[0], wavenumbers_473, norm_473, peak_wavenumbers_473, peak_heights_473, snr_473, apc.aegean, 'a.'),
    (axes[1], wavenumbers_532, norm_532, peak_wavenumbers_532, peak_heights_532, snr_532, apc.lime,   'b.'),
    (axes[2], wavenumbers_785, norm_785, peak_wavenumbers_785, peak_heights_785, snr_785, apc.dragon, 'c.'),
]

snr_all = np.concatenate([snr_473, snr_532, snr_785])
snr_min, snr_max = snr_all.min(), snr_all.max()
norm_snr = mcolors.Normalize(vmin=snr_min, vmax=snr_max)
cmap_snr = apc.gradients.viridis.to_mpl_cmap()

for j, (ax, wns, spectrum, peak_wns, peak_hs, snrs, line_color, label) in enumerate(datasets):
    ax.plot(wns, spectrum, color=line_color, linewidth=1.2, alpha=0.8)
    sc = ax.scatter(peak_wns, peak_hs, c=snrs, cmap=cmap_snr, norm=norm_snr,
                    s=40, zorder=5, edgecolors='k', linewidths=0.4)
    cbar = fig.colorbar(sc, ax=ax, fraction=0.02, pad=0.01)
    cbar.set_label('SNR', fontsize=8)
    ax.set_title(label, loc='left', fontweight='bold')
    ax.set_facecolor(apc.parchment)
    if j == 2:
        ax.set_ylabel('normalized intensity')
        ax.set_xlabel('Raman Shift (cm$^{-1}$)')
apc.mpl.style_plot(monospaced_axes="both")
plt.show()
```

## Spatial heterogeneity analysis

All three datasets are compared in the univariate marker maps below; for PCA and NMF we focus exclusively on the 473 nm dataset, which has the highest point density and SNR of the three (see justification after the marker maps). In all datasets, it is possible to distinguish the cells from the background (to varying degrees). However, we're interested in if we can distinguish cellular substructures or organelles, such as the eyespot, chloroplasts, or starch granules. There are a few ways we can start to pull apart the data to find natural clusters or decompose the spectra themselves. Not all methods are equally useful for all the datasets, so it's worth considering what to try at the outset. 

At the most basic level, we could do simple univariate analysis by creating maps that show intensity at a single wavenumber for each xy point. We could select known key Raman peak positions as markers for specific biomolecules relevant to our system, for instance ~1530 cm^-1^ for the C=C symmetrical stretch of carotenoids, which we know are present in *Chlamydomonas*. In each map, we would expect that hotspots of high intensity at this wavenumber would correlate to the positions of organelles or components rich in the biomolecule of interest. For carotenoids, this could be the eyespot. This is something we can already do to an extent with the interactive spectral maps above. However, applying this on baseline-corrected and background-subtracted, masked, data could be more useful. This method is simple and biologically grounded, but requires prior knowledge of what to look for and the wavenumber of interest, cannot resolve overlapping markers, and might not be great for multi-molecular organelles. It is commonly done in biological samples analyzed with Raman spectroscopy (cite).

Next, we could try various forms of unsupervised decomposition to find spectrally distinct components. Principal Component Analysis (PCA) finds directions of maximum spectral variance, useful for quick compartmentalization into different composition groups (e.g. eyespot from non-eyespot) and general exploration of the dataset. However, it cannot always separate mixed components, might miss subtle differences, and can be difficult to interpret since the components can be negative. It is often a first pass analysis for spectral data (cite). Non-negative Matrix Factorization (NMF), by contrast, decomposes the data into non-negative spectral components. The components are non-negative and more likely interpretable, it estimates unmixing, and can be effective for biochemical compartmentalization (cite). However, it requires selecting the number of components and requires moderate SNR at least. Other approaches — including hierarchical clustering, k-means, and UMAP — are applicable to this type of data but were not pursued here. Our dataset is small and lacks sufficient point density per cell for clustering to be reliable, and UMAP would require pairing with a decomposition method to be interpretable. Citations for those approaches are provided for reference [cite, cite, cite].


### Single-wavenumber marker maps

Let's select four different molecules of interest: carotenoids, chlorophylls, starches, and lipids. All of these should be abundant in the cells, and have Raman-detectable vibrational modes. We can use the C=C stretching mode at ~1530 cm^-1^ as a marker for carotenoids, the pyrrole mode at  ~746 cm^-1^ for chlorophylls, the pyranose ring mode at ~478 cm^-1^ for starch, and ~2911 cm^-1^ for lipids and other molecules with C-H stretching modes. These numbers can also be changed in the code cell below.

It's important to note that these are not the only peaks associated with any of these molecules, nor are these molecules the only ones that can have peaks at those wavenumbers. However, given what we know of the biology of these cells and prior work, they are reasonable Raman markers to use (cite: Ji et al 2014, others).  

```python
# Phase 2 Step 1: Extract Baseline-Corrected Intensity at Target Wavenumbers

import numpy as np
import pandas as pd
from pybaselines.whittaker import airpls

# IMPORTANT: Store the coordinate arrays that match the intensity arrays
# These should match the masked datasets used for baseline correction
coords_for_markers = {
    '473nm': blue_xy[mask_473].reset_index(drop=True),
    '532nm': green_xy[mask_532].reset_index(drop=True),
    '785nm': red_xy[mask_785].reset_index(drop=True)
}

print("Coordinate arrays for markers:")
print(f"  473nm: {len(coords_for_markers['473nm'])} points")
print(f"  532nm: {len(coords_for_markers['532nm'])} points")
print(f"  785nm: {len(coords_for_markers['785nm'])} points")
print()

# Define target wavenumbers for each marker
markers = {
    'Carotenoid': 1530,
    'Chlorophyll': 746,
    'Starch': 478,
    'Lipids-plus': 2911
}

# Function to apply baseline correction and extract marker intensity
def extract_marker_intensity(spectra_array, wavenumbers, target_wn, lam=1e6):
    """
    Apply baseline correction to each spectrum and extract intensity at target wavenumber.

    Parameters:
    -----------
    spectra_array : DataFrame
        Spectral intensity array (n_points × n_wavenumbers)
    wavenumbers : array
        Wavenumber values
    target_wn : float
        Target wavenumber for marker
    lam : float
        airPLS smoothing parameter

    Returns:
    --------
    marker_intensities : array
        Baseline-corrected intensity at target wavenumber for each point
    """
    # Find closest wavenumber index
    idx = np.argmin(np.abs(wavenumbers - target_wn))
    actual_wn = wavenumbers[idx]
    print(f"  Target: {target_wn} cm⁻¹ → Actual: {actual_wn:.1f} cm⁻¹ (index {idx})")

    # Apply baseline correction to each spectrum
    marker_intensities = np.zeros(len(spectra_array))
    for i, spectrum in enumerate(spectra_array.values):
        baseline_i = airpls(spectrum, lam=lam)
        baseline_i = baseline_i[0] if isinstance(baseline_i, tuple) else baseline_i
        corrected_spectrum = spectrum - baseline_i
        marker_intensities[i] = corrected_spectrum[idx]

    return marker_intensities

# Extract marker intensities for all wavelengths
marker_data = {}

for marker_name, target_wn in markers.items():
    print(f"{marker_name} ({target_wn} cm⁻¹):")
    marker_data[marker_name] = {}

    # 473 nm
    if target_wn >= wavenumbers_473.min() and target_wn <= wavenumbers_473.max():
        print("  473 nm:")
        marker_data[marker_name]['473nm'] = extract_marker_intensity(
            blue_intensity_masked, wavenumbers_473, target_wn
        )
    else:
        print(f"  473 nm: Out of range (range: {wavenumbers_473.min():.0f}-{wavenumbers_473.max():.0f} cm⁻¹)")
        marker_data[marker_name]['473nm'] = None

    # 532 nm
    if target_wn >= wavenumbers_532.min() and target_wn <= wavenumbers_532.max():
        print("  532 nm:")
        marker_data[marker_name]['532nm'] = extract_marker_intensity(
            green_intensity_masked, wavenumbers_532, target_wn
        )
    else:
        print(f"  532 nm: Out of range (range: {wavenumbers_532.min():.0f}-{wavenumbers_532.max():.0f} cm⁻¹)")
        marker_data[marker_name]['532nm'] = None

    # 785 nm (requires background subtraction first)
    if target_wn >= wavenumbers_785.min() and target_wn <= wavenumbers_785.max():
        print("  785 nm:")
        # Subtract background before baseline correction
        red_intensity_bkgd_sub = red_intensity_masked.values - mean_background_785.values
        marker_data[marker_name]['785nm'] = extract_marker_intensity(
            pd.DataFrame(red_intensity_bkgd_sub, columns=red_intensity_masked.columns),
            wavenumbers_785, target_wn
        )
    else:
        print(f"  785 nm: Out of range (range: {wavenumbers_785.min():.0f}-{wavenumbers_785.max():.0f} cm⁻¹)")
        marker_data[marker_name]['785nm'] = None

    print()
```

### PCA on masked, baseline-corrected data

To make a fair comparison between PCA and NMF, we re-run PCA on the same input used for NMF: the masked, baseline-corrected dataset. This removes the cell/background contrast that dominated PC1 in the unmasked analysis, giving PCA the best possible chance to resolve within-cell variation.


```python
# | code-fold: true
# | code-summary: "Source code for preprocessing pipeline (baseline + L2 normalization):"

from pybaselines.whittaker import airpls

# Apply airPLS baseline correction to every spectrum in the masked 473 nm dataset,
# then clip residual negatives to zero.
# This removes per-spectrum fluorescence background before PCA/NMF while
# preserving absolute intensity differences between pixels — intensity carries
# biological information here (e.g. resonance-enhanced eyespot carotenoids).

spectra_raw = blue_intensity_masked.values.copy()
n_spectra, n_wn = spectra_raw.shape
X_473_corrected = np.zeros_like(spectra_raw)

print(f"Running airPLS on {n_spectra} spectra...")
for i, spectrum in enumerate(spectra_raw):
    baseline, _ = airpls(spectrum, lam=1e6)
    X_473_corrected[i] = spectrum - baseline
    if (i + 1) % 200 == 0:
        print(f"  {i + 1}/{n_spectra}")

# Clip small negatives left by baseline subtraction
X_473_corrected = np.clip(X_473_corrected, 0, None)

X_473_processed = X_473_corrected

print(f"Done. Output shape: {X_473_processed.shape}")
print(f"Value range: [{X_473_processed.min():.4f}, {X_473_processed.max():.4f}]")
```

### NMF: Spectral decomposition into interpretable components

NMF (non-negative matrix factorization) decomposes the spectral data matrix into a set of non-negative component spectra and corresponding spatial abundance maps. Because Raman intensities are non-negative, this constraint is physically meaningful: components represent additive contributions of distinct chemical signatures. This makes NMF components more directly interpretable than PCA loadings, which can be negative and represent variance contrasts rather than physical mixtures.

Unlike PCA above, which was applied to unmasked data to separate cells from background, NMF is applied to the masked, baseline-corrected dataset — the goal here is characterizing within-cell chemical variation, not cell detection. We applied NMF to the masked 473 nm dataset (804 on-cell points × 1373 wavenumbers) using 10 components. We tested 1–15 components; reconstruction quality exceeded 99.7% throughout with no clear inflection point, so the number of components cannot be selected from reconstruction error alone. We used 10 and evaluated components directly by inspecting their spectra and spatial maps. Unlike PCA, NMF has no inherent ordering — components are not ranked by explained variance. We present all 10 component spectra and spatial maps below, using spectral interpretability and spatial coherence as evaluation criteria.


```python
# | code-fold: true
# | code-summary: "Source code for NMF top-10 components (spectra and spatial maps):"
# | label: fig-473-nmf-components
# | fig-cap: "NMF components 1-10 for masked baseline-corrected 473 nm dataset. Left: component spectrum; right: spatial abundance map."

n_show = 10
fig, axes = plt.subplots(n_show, 2, figsize=(18, 40), layout='constrained',
                         facecolor=apc.parchment,
                         gridspec_kw={'width_ratios': [3, 1]})

for i in range(n_show):
    ax_l = axes[i, 0]
    ax_l.plot(wavenumbers_473, nmf_components_473[i], color=apc_473, linewidth=1)
    ax_l.set_xlim(wavenumbers_473[0], wavenumbers_473[-1])
    ax_l.set_facecolor(apc.parchment)
    ax_l.grid(False)
    if i == n_show - 1:
        ax_l.set_ylabel('intensity')
        ax_l.set_xlabel('Raman shift (cm$^{-1}$)')

    ax_s = axes[i, 1]
    scatter = ax_s.scatter(
        blue_xy_masked['x_pos'], blue_xy_masked['y_pos'],
        c=nmf_scores_473[:, i],
        cmap=apc.gradients.red_blue.to_mpl_cmap(),
        s=20, alpha=0.7
    )
    plt.colorbar(scatter, ax=ax_s, fraction=0.046, pad=0.04)
    ax_s.set_aspect('equal')
    ax_s.set_facecolor(apc.parchment)
    if i == n_show - 1:
        ax_s.set_xlabel('X (μm)')

apc.mpl.style_plot(monospaced_axes='both')
plt.show()
```

Examining the 15 components, most are dominated by broad high-wavenumber features or scattered noise, with flat spatial distributions that carry no meaningful biological information. Two components stand out.

**Component 10** has the most coherent, interpretable spectrum: sharp peaks in the fingerprint region at ~1157 and ~1531 cm⁻¹, matching the ν₂ (C-C stretch) and ν₁ (C=C stretch) modes of carotenoids, alongside broader features consistent with chlorophyll and lipid contributions. Its spatial map shows signal distributed across all four cells with relatively uniform coverage, consistent with a bulk cell signal integrating pigmentation from the chloroplast and general cellular composition.

**Component 2** has a narrower spectral profile dominated by the same carotenoid doublet (~1162 and ~1533 cm⁻¹) but with minimal contribution elsewhere. Its spatial map is highly localized — only a few bright points per cell — in contrast to the diffuse pattern of component 10. This combination of a pure carotenoid spectrum and concentrated spatial distribution is consistent with the eyespot, a β-carotene-rich organelle known to produce a strong resonance-enhanced Raman signal under 473 nm excitation [cite].

The key advantage over the single-wavenumber marker maps above is that both components contribute to intensity at ~1530 cm⁻¹, but NMF separates them because they differ in their broader spectral shape and spatial distribution. Univariate mapping at that wavenumber alone cannot distinguish a localized carotenoid-rich structure from diffuse chloroplast pigmentation.

This interpretation is tentative: confirming the eyespot assignment would require a co-registered brightfield or fluorescence image of the same cells. The beam size (~1 μm) is comparable to the eyespot diameter, so a single bright point in the spatial map could reflect a genuine localized structure or local intensity variation. Nonetheless, the separation of a spectrally distinct, spatially concentrated component from a diffuse bulk signal illustrates what NMF can offer for exploratory analysis of Raman map data.


```python
# | code-fold: true
# | code-summary: "Source code for NMF component analysis:"

from scipy.signal import find_peaks

# Define literature reference peaks for biological interpretation
literature_peaks = {
    'Carotenoid': [1008, 1155, 1530],  # ν₃, ν₂, ν₁ (strongest)
    'Chlorophyll_a': [747, 1207, 1325, 1548],
    'Chlorophyll_b': [1220, 1340, 1560],
    'Lipid': [1440, 1460, 2850, 2950],  # CH₂, C-H stretch
    'Starch': [1080, 1440],  # C-O-C, CH₂
    'Protein': [1003, 1650, 1680]  # Phe, Amide I
}

import pandas as pd

# Spatial localization is measured by the coefficient of variation (CV) of each
# component's scores across all map points: CV = std / mean. A high CV (>0.5)
# means most points have near-zero weight and only a few are strongly activated —
# i.e., the component is concentrated in a small region. A low CV means the
# component contributes roughly equally everywhere (a bulk/background feature).
# This gives a fast numeric filter across all 10 components before visual inspection.

rows = []
for i in range(n_components_nmf):
    scores = nmf_scores_473[:, i]
    cv = np.std(scores) / (np.mean(scores) + 1e-10)

    component_spectrum = nmf_components_473[i, :]
    norm_component = component_spectrum / component_spectrum.max()
    comp_peaks, comp_props = find_peaks(norm_component, prominence=0.05, height=0.02, distance=5)
    comp_peak_wns = wavenumbers_473[comp_peaks]
    top_idx = np.argsort(comp_props['prominences'])[::-1][:5]
    top_peaks = ', '.join([f'{comp_peak_wns[j]:.0f}' for j in top_idx]) if len(comp_peaks) > 0 else '—'

    lit_hits = []
    max_int = component_spectrum.max()
    for category, lit_peaks in literature_peaks.items():
        for lp in lit_peaks:
            if lp < wavenumbers_473.min() or lp > wavenumbers_473.max():
                continue
            if component_spectrum[np.argmin(np.abs(wavenumbers_473 - lp))] / max_int > 0.30:
                lit_hits.append(category.replace('_', ' '))
                break
    lit_str = ', '.join(dict.fromkeys(lit_hits)) if lit_hits else '—'

    rows.append({'Component': i + 1, 'CV': round(cv, 2),
                 'Localized (CV>0.5)': '✓' if cv > 0.5 else '',
                 'Top peaks (cm⁻¹)': top_peaks, 'Literature match': lit_str})

df = pd.DataFrame(rows).set_index('Component')
print(df.to_string())
```

### Comparing approaches: univariate, PCA, and NMF

# Discussion

## What we found

Across three excitation wavelengths, spontaneous Raman mapping of live *Chlamydomonas* cells consistently produced detectable signals from carotenoid pigments, with the 473 nm dataset yielding the most interpretable data. The dominance of the carotenoid ν₁ and ν₂ modes (~1531 and ~1157 cm⁻¹) across all three wavelengths reflects both their abundance in *Chlamydomonas* and the resonance enhancement predicted by the UV-Vis spectra — particularly strong under 473 nm excitation. Chlorophyll peaks were detectable but not resonantly enhanced, consistent with none of the excitation lines overlapping the Soret or Q bands. The starch marker at ~478 cm⁻¹ was not detectable under any condition, likely reflecting the limits of acquisition parameters on a demo dataset rather than absence of starch.

The most informative result came from applying NMF to the 473 nm dataset. Two components stood out: a diffuse bulk cell signal (component 10) with a mixed pigment spectrum consistent with the chloroplast, and a spatially localized component (component 2) with a narrow carotenoid spectral profile consistent with the eyespot. This separation — of two structures that both contribute to the same Raman peak at ~1530 cm⁻¹ — illustrates the key advantage of combining spatial and spectral information: univariate mapping at a single wavenumber cannot distinguish a concentrated organelle from diffuse pigmentation distributed across the cell.

## How this differs from bulk Raman

Our companion publication compared spontaneous Raman spectrometers using bulk spectra of *Chlamydomonas* and other samples, providing a framework for instrument selection [cite]. That work treats spectra as population-level averages. Confocal mapping adds a spatial dimension that bulk acquisition loses entirely. Even a single-cell map with ~800 on-cell points contains more structural information than any number of bulk spectra — the question shifts from *what peaks are present* to *where within the cell* they are concentrated. This makes mapping particularly well-suited for studying organelle-level composition, cell-to-cell variation, and responses to perturbation that may be localized rather than cell-wide.

## What worked and what didn't

The 473 nm dataset was clearly the most useful for spatial analysis. In hindsight, all three acquisitions should have used the same instrument, species, and acquisition parameters for a controlled comparison. The 532 nm dataset suffered from focus issues that reduced usable point density. The 785 nm dataset was acquired on a different instrument and species, making cross-wavelength comparisons confounded. For a rigorous comparison of excitation wavelengths in mapping mode, these would need to be re-acquired under matched conditions.

The analysis pipeline — masking, baseline correction, univariate marker maps, PCA, NMF — reflects a logical progression from simple to complex and from supervised (marker-guided) to unsupervised (data-driven). In practice, PCA was only useful for cell/background separation; NMF was the more productive tool for within-cell analysis. For datasets with higher point density per cell or cleaner signal, clustering methods such as hierarchical clustering or k-means, and manifold learning approaches such as UMAP, would be natural next steps.

## What to do next

To make stronger biological claims, the key improvements would be: (1) acquire mapping data on the same species and instrument across all wavelengths, ideally with a co-registered brightfield image to validate organelle assignments; (2) increase map density so that individual organelles are sampled by multiple points; (3) include biological replicates across multiple cells and acquisition days. With those improvements, this type of analysis could detect changes in pigment composition or organelle distribution in response to growth conditions, stress, or genetic perturbation — making spontaneous Raman mapping a practical single-cell phenotyping tool for *Chlamydomonas* and related organisms.

## Caveats

Our aim was to evaluate the instruments and analysis methods, rather than conduct a controlled scientific investigation. Therefore, there are significant caveats to this dataset that preclude drawing strong scientific claims. We used the same species for the 473 nm and 532 nm excitation acquisitions, which were acquired on the same instrument on the same day. However, the focusing in z was different between the two acquisitions and they were on different cell clusters. The 785 nm data was acquired on a different instrument and species, on a different day, and we focused on a single cell rather than a cluster. Given that our goal was single-cell analysis to evaluate subcellular structure, we did detailed scans on few cells overall, so there is a low n. We also did not perform this on biological or technical replicates or across days.
